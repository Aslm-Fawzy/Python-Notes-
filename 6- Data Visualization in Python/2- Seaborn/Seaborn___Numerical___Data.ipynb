## **Histplots :-**

```python
sns.histplot(df_tips['total_bill'], color='r')
```

```python
sns.histplot(df_tips['total_bill'], color='r',bins=20)
```

```python
sns.histplot(data=df_tips,x="total_bill",color="r",stat="percent")
'''
stat is Aggregate statistic to compute in each bin and can be : 
   1. count: show the number of observations in each bin (defualt)
   2. frequency: show the number of observations divided by the bin width
   3. probability: or proportion: normalize such that bar heights sum to 1
   4. percent: normalize such that bar heights sum to 100
   5. density: normalize such that the total area of the histogram equals 1
'''
```

```python
sns.histplot(data=df_tips,x="total_bill",color="r",shrink=.8)
```

```python
sns.histplot(data=df_tips,x="total_bill",color="r",element="poly")
#element can be {bars, step, poly}
```

```python
sns.histplot(data=df_tips,x="total_bill",hue="sex",element="poly")
```

## **Scatterplots :-**

```python
sns.scatterplot(x='total_bill',y='tip',data=df_tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df_tips,color='k')
```

```python
sns.scatterplot(data=df_tips, x="total_bill", y="tip", hue="time")
```

## **Regressionplots :-**

```python
sns.lmplot(x="total_bill", y="tip", data=df_tips,height=5,aspect=1.6)
```

```python
sns.lmplot(x="total_bill", y="tip", data=df_tips,height=5,aspect=1.6,hue='smoker')
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df_tips
              ,height=5,aspect=1.6, markers=["o", "x"])
```

```python
sns.lmplot(x="total_bill", y="tip", col="smoker"
           ,height=5,aspect=1.6,data=df_tips)
```

```python
sns.lmplot(x="total_bill", y="tip", col="time"
           ,height=5,aspect=1.4,data=df_tips)
```

```python
sns.lmplot(x="total_bill", y="tip", col="smoker",row="time"
           ,height=4,aspect=1.2,data=df_tips)
```

## **Joinplots :-**


> Measure the relation , correlation and dependency between two vairiables only  






```python
sns.jointplot(x='total_bill',y='tip',data=df_tips,height=7,color='k') 
#no aspect in joinplot there is height only
```

```python
sns.jointplot(x='total_bill',y='tip',data=df_tips,height=7,
              color='k',marginal_ticks=True) 
```

```python
sns.jointplot(x='total_bill',y='tip',data=df_tips,height=7,color='k',
              marginal_ticks=True,marker='X') 
```

```python
#defult kind is scatter  
sns.jointplot(x='total_bill',y='tip',data=df_tips,height=7,color='k'
              ,kind='reg',marginal_ticks=True)
```

```python
sns.jointplot(x='total_bill', y='tip', data=df_tips, kind='kde',
              height=8, color='k',marginal_ticks=True)
```
