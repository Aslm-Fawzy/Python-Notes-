# SQLite 
SQLite is a C-language library that implements a small, fast, self-contained, high-reliability, full-featured, SQL database engine. SQLite is the most used database engine in the world. SQLite is built into all mobile phones and most computers and comes bundled inside countless other applications that people use every day. 
# Python SQLite3 
 Python SQLite3 module is used to integrate the SQLite database with Python. It is a standardized Python DBI API 2.0 and provides a straightforward and simple-to-use interface for interacting with SQLite databases.

```python
import sqlite3
```

# Read Data 

```python
# Create Connection 
db = sqlite3.connect('books.db')
```

```python
# Setting up Cursor 
cr = db.cursor()
```

```python
#Return Table of Categories
cr.execute('SELECT * FROM categories').fetchall()
```

```python
#Return Table of Categories and Ordering it
cr.execute('SELECT * FROM categories ORDER BY name ').fetchall()
```

```python
#Return frist 5 rows from Table of Categories and Ordering it
cr.execute('SELECT * FROM categories ORDER BY name LIMIT 5 ').fetchall()
```

```python
#Return Table of books
cr.execute('SELECT * FROM books').fetchall()
```

```python
#Return books with price less than 20 
sql = 'SELECT * FROM books WHERE price < ?'
data = [20]
cr.execute(sql,data).fetchall()
```

```python
#Return books with price less than 50 and rate more than 3 
sql = 'SELECT * FROM books WHERE price < ? AND rate > ?'
data = [50,3]
rows = cr.execute(sql,data).fetchall()
rows
```

```python
for bookid,bookname,price,rate,category  in rows:
    print(f'user with id {bookid} has name {bookname} with price {price} , rate {rate} and its category_id = {category}')
    print('-'*100)
```

```python
# Close Connection
db.close()
```
